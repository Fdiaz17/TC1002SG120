1, The top10 country with most Vaccination Percentage

```python

df = pd.DataFrame(data)
df.loc[:,['country', 'total_vaccinations_per_hundred']].groupby('country').max().sort_values(by="total_vaccinations_per_hundred",ascending=False).head(10)
```

2, The top10 country with least Vaccination Percentage

```python
df = pd.DataFrame(data)
df.loc[:,['country', 'total_vaccinations_per_hundred']].groupby('country').max().sort_values(by="total_vaccinations_per_hundred",ascending=True).head(10)
```

3, The top10 country with most Daily Vaccination Rate

```python

df = pd.DataFrame(data)
df.loc[:,['country', 'daily_vaccinations_per_million']].groupby('country').max().sort_values(by="daily_vaccinations_per_million",ascending=False).head(10)
```

4, The top10 country with least Daily Vaccination Rate

```python
df = pd.DataFrame(data)
df.loc[:,['country', 'daily_vaccinations_per_million']].groupby('country').max().sort_values(by="daily_vaccinations_per_million",ascending=True).head(10)
```

5, Stand Distribution of Vaccination Percentage

```python
(df.loc[:,['country', 'total_vaccinations_per_hundred']].groupby('country').max())['total_vaccinations_per_hundred'].std()
```

```python
(df.loc[:,['country', 'total_vaccinations_per_hundred']].groupby('country').max())['total_vaccinations_per_hundred'].mean()
```
