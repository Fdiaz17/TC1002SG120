# Filtro de atletas que obtuvieron al menos una medalla

```python
filtro_athletes_IMC=athletes_IMC[(athletes_IMC['gold']>0) | (athletes_IMC['silver']>0) | (athletes_IMC['bronze']>0)]
filtro_athletes_IMC
```

# Mapa de calor de los atletas que tuvieron al menos una medalla, comparando su IMC,peso, altura y la medalla

```python
colormap=plt.cm.viridis
plt.figure(figsize=(12,12))
plt.title('Pearson map', y=1.05,size=15)
sb.heatmap(filtro_athletes_IMC.iloc[:, [5,6,8,9,10,11]].astype(float).corr(),linewidths=0.1,vmax=1.0,square=True,cmap=colormap,linecolor='white',annot=True)
```
