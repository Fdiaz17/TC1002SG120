```python
 import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import sklearn
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import scale
import sklearn.metrics as sm
from sklearn import datasets
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
X = np.array(df[['daily_vaccinations_per_million','total_vaccinations_per_hundred','people_vaccinated_per_hundred']])
Y = np.array(df['people_fully_vaccinated_per_hundred'])
X.shape
```

```python
Y.shape
```

```python
X[:,0] = [0 if math.isnan(x) else x for x in X[:,0]]
X[:,1] = [0 if math.isnan(x) else x for x in X[:,1]]
X[:,2] = [0 if math.isnan(x) else x for x in X[:,2]]
```

```python
Y.min()
```

```python
import math
Y = [0 if math.isnan(y) else math.ceil(y) for y in Y]
```

```python
min(Y)
```

```python
max(Y)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
colores=['blue','red','green', 'blue', 'cyan', 'yellow','orange', 'black', 'pink', 'brown', 'purple']
asignar=[]
for row in Y:
    asignar.append(colores[row//10])
ax.scatter(X[:,0], X[:,1], X[:,2], c=asignar, s=60);
```

```python
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
score
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5).fit(X)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
# Obtenemos las etiquetas de cada punto de nuestros datos
labels = kmeans.predict(X)
# Obtenemos los centroids
C = kmeans.cluster_centers_
colores=['red','green','blue','cyan','yellow']
asignar=[]
for row in labels:
    asignar.append(colores[row])
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=60)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c=colores, s=1000);
```

```python
# Hacemos una proyecciÃ³n a 2D con los diversos ejes
f1 = X[:, 0]
f2 = X[:, 1]
plt.scatter(f1, f2, c=asignar, s=70)
plt.scatter(C[:, 0], C[:, 1], marker='*', c=colores, s=1000)
plt.show()
```

```python
 # Hacemos una proyecciÃ³n a 2D con los diversos ejes
f1 = X[:, 0]
f2 = X[:, 2]
plt.scatter(f1, f2, c=asignar, s=70)
plt.scatter(C[:, 0], C[:, 2], marker='*', c=colores, s=1000)
plt.show()
```
