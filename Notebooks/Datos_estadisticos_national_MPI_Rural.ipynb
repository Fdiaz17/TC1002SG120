```python
X = np.array(dataframe[["Headcount Ratio Rural","MPI Rural","Intensity of Deprivation Rural"]])
y = np.array([x for x in range(len(dataframe.index))])
X.shape
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0],X[:,1],X[:,2], s=60);
```

```python
Nc = range(1,20)
kmeans =[KMeans(n_clusters=i)for i in Nc]
kmeans
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
score
plt.plot(Nc,score)
plt.xlabel('Number of clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
kmeans= KMeans(n_clusters=5).fit(X)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
labels = kmeans.predict(X)
C = kmeans.cluster_centers_
colores=['red','green','blue','cyan','yellow']
asignar=[]
for row in labels:
    asignar.append(colores[row])

fig=plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:,0],X[:,1],X[:,2], c=asignar, s=60)
ax.scatter(C[:,0],C[:,1],C[:,2], marker='*', c=colores, s=1000);
```

```python
f1= dataframe["Intensity of Deprivation Urban"].values
f2= dataframe['MPI Rural'].values

plt.scatter(f1,f2,c=asignar,s=70)
plt.scatter(C[:,0],C[:,1], marker='*', c=colores, s=1000)
plt.show()
```

```python
print (classification_report(y, labels));
```
